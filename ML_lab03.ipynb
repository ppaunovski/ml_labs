```python
import pandas as pd
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
mice_protein_expression = fetch_ucirepo(id=342)

# data (as pandas dataframes)
X = mice_protein_expression.data.features
y = mice_protein_expression.data.targets

# metadata
print(mice_protein_expression.metadata)

# variable information
print(mice_protein_expression.variables)
```

```python
X
```

```python
y
```

```python
df = X.drop(columns = ['Genotype', 'Treatment', 'Behavior'])
df
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

mice = IterativeImputer(random_state=42)

mice.fit(df)
df_imputed = mice.transform(df)
len(df_imputed)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
df_pca = pca.fit_transform(df_imputed)
```

```python
import matplotlib.pyplot as plt


plt.scatter(x=df_pca[:,0], y=df_pca[:,1])
plt.show()
```

```python
pca = PCA(n_components=3)
df_pca_3d = pca.fit_transform(df_imputed)
```

```python
import plotly.express as px
px.scatter_3d(x=df_pca_3d[:,0], y=df_pca_3d[:,1], z=df_pca_3d[:,2])
```

```python
from sklearn.cluster import KMeans
```

```python
scores = []

for i in range(1, 50, 2):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(df_imputed)
  scores.append(kmeans.inertia_)

scores
```

```python
plt.plot(range(1,50,2), scores, marker='x')
plt.xticks(range(1,50,2))
plt.show()
```

```python
from sklearn.manifold import TSNE
k=7
kmeans = KMeans(n_clusters=7, n_init='auto')
kmeans.fit(df_imputed)
kmeans_labels = kmeans.labels_

tsne = TSNE(n_components=3)
tsne_data = tsne.fit_transform(df_imputed)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(tsne_data[:, 0], tsne_data[:, 1], tsne_data[:, 2], c=kmeans_labels)
plt.show()

tsne = TSNE(n_components=2)
tsne_data = tsne.fit_transform(df_imputed)


plt.scatter(tsne_data[:, 0], tsne_data[:, 1], c=kmeans_labels)

plt.show()
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=7)
gmm_labels = gmm.fit_predict(df_imputed)


tsne = TSNE(n_components=3)
tsne_data = tsne.fit_transform(df_imputed)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(tsne_data[:, 0], tsne_data[:, 1], tsne_data[:, 2], c=gmm_labels)
plt.show()

tsne = TSNE(n_components=2)
tsne_data = tsne.fit_transform(df_imputed)


plt.scatter(tsne_data[:, 0], tsne_data[:, 1], c=gmm_labels)

plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import silhouette_score
import numpy as np
```

```python
dbscan = DBSCAN(eps=0.8).fit(df_imputed)
dbscan_labels = dbscan.labels_

tsne = TSNE(n_components=3)
tsne_data = tsne.fit_transform(df_imputed)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(tsne_data[:, 0], tsne_data[:, 1], tsne_data[:, 2], c=dbscan_labels)
plt.show()

tsne = TSNE(n_components=2)
tsne_data = tsne.fit_transform(df_imputed)


plt.scatter(tsne_data[:, 0], tsne_data[:, 1], c=dbscan_labels)

plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
plt.figure(figsize=(12, 6))
dendrogram(linkage(df_imputed, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=4)
agg.fit(df_imputed)
agg_labels = agg.labels_

tsne = TSNE(n_components=3)
tsne_data = tsne.fit_transform(df_imputed)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(tsne_data[:, 0], tsne_data[:, 1], tsne_data[:, 2], c=agg_labels)
plt.show()

tsne = TSNE(n_components=2)
tsne_data = tsne.fit_transform(df_imputed)


plt.scatter(tsne_data[:, 0], tsne_data[:, 1], c=agg_labels)

plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score
```

```python
print(f'kmeans CH index: {calinski_harabasz_score(df_imputed, kmeans_labels)}')
print(f'Gaussian Mixture CH index: {calinski_harabasz_score(df_imputed, gmm_labels)}')
print(f'DBSCAN CH index: {calinski_harabasz_score(df_imputed, dbscan_labels)}')
print(f'Agglomerative clustering CH index: {calinski_harabasz_score(df_imputed, agg_labels)}')
```
